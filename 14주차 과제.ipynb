```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
import scipy as sp
import pandas as pd
import sklearn as sk
import scipy.stats as stats
from scipy.stats import norm
```

```python
np.random.seed(0)
N1 = 200
N2 = 300
x1 = sp.stats.norm(0, 1).rvs(N1)
x2 = sp.stats.norm(1.5, 1.5).rvs(N2)
ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label="N1 data set")
ax = sns.distplot(x2, kde=False, fit=sp.stats.norm, label="N2 data set")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```

```python
sp.stats.ks_2samp(x1, x2)
```
