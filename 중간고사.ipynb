```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

mu = [0.3, 0.25, 0.2, 0.15, 0.1]
rv = sp.stats.multinomial(1, mu)

np.random.seed(1)
x = rv.rvs(100)
y = x.sum(axis = 0) / float(len(x))

xx = np.arange(1, 6)
xx_ohe = pd.get_dummies(xx)

df = pd.DataFrame({'Theory': rv.pmf(xx_ohe.values), 'Simulation': y}, index=np.arange(1, 6)).stack()
df = df.reset_index()
df.columns = ['sample value', 'type', 'ratio']
df.pivot('sample value', 'type', 'ratio')
df

sns.barplot(x='sample value', y='ratio', hue='type', data=df)
plt.title("simulation dist. and theoretical dist. of catagorical dist.")
plt.show()
```

```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

mu = [0.3, 0.25, 0.2, 0.15, 0.1]
rv = sp.stats.multinomial(1, mu)

xx = np.arange(1, 6)
print(xx)
xx_ohe = pd.get_dummies(xx)
print(xx_ohe)
plt.bar(xx, rv.pmf(xx_ohe.values))
plt.ylabel("P(x)")
plt.xlabel("sample values")
plt.title("pmf of categorical distribution")
plt.show()
```

```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from numpy import linspace

np.random.seed(1)
plt.figure(figsize=(7, 10))
for i, N in enumerate([2, 3, 4]):
    X = np.random.rand(1000, N)
    Xbar = (X.mean(axis=1) - 3) * np.sqrt(2 * N)
    ax = plt.subplot(3, 2, 2 * i + 1)
    sns.distplot(Xbar, bins=10, kde=False, norm_hist=True)
    plt.xlim(-10, 10)
    plt.yticks([])
    ax.set_title("N = {0}".format(N))
    plt.subplot(3, 2, 2 * i + 2)
    sp.stats.probplot(Xbar, plot=plt)

plt.tight_layout()
plt.show()
```
