```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

mu = [0.25, 0.25, 0.25, 0.25]
rv = sp.stats.multinomial(1, mu)

np.random.seed(1)
x = rv.rvs(10)
y = x.sum(axis = 0) / float(len(x))

xx = np.arange(1, 5)
xx_ohe = pd.get_dummies(xx)

df = pd.DataFrame({'Theory': rv.pmf(xx_ohe.values), 'Simulation': y}, index=np.arange(1, 5)).stack()
df = df.reset_index()
df.columns = ['sample value', 'type', 'ratio']
df.pivot('sample value', 'type', 'ratio')
df

sns.barplot(x='sample value', y='ratio', hue='type', data=df)
plt.title("simulation dist. and theoretical dist. of catagorical dist.")
plt.show()
```
