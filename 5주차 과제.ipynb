```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
n=5
mu=0.5
rv=sp.stats.binom(n, mu)

x=rv.rvs(10, random_state=0)
y=np.mean(x)
print(y)

z=np.var(x, ddof=1)
print(z)

s=sp.stats.describe(x)
print(s)
```

```python
rv=sp.stats.norm(loc=0.5, scale=5)
xx=np.linspace(-6, 6, 10)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title("probability density function")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()
```

```python
rv=sp.stats.norm(loc=0.5, scale=5)
rv.rvs(size=(3,5), random_state=0)
```

```python
rv=sp.stats.norm(loc=0.5, scale=5)
sns.displot(rv.rvs(size=10, random_state=0))
plt.title("Result of Random Sample Generation")
plt.xlabel("Sample Value")
plt.ylabel("Count")
plt.xlim(-6, 6)
plt.show()
```

```python
np.random.seed(0)
n=5
mu=0.5
rv=sp.stats.binom(n, mu)
x=rv.rvs(10)
sns.countplot(x)
plt.title("Simulation Result")
plt.xlabel("Sample Value")
plt.show()
```
