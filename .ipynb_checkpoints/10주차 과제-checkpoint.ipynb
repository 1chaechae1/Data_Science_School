```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt

xx = np.linspace(0, 1, 1000)
plt.subplot(221)
plt.fill_between(xx, sp.stats.beta(3, 2).pdf(xx))
plt.ylim(0, 6)
plt.title("a=3 b=2 MODE={0}          ".format((3-1)/(3+2-2)))
plt.subplot(222)
plt.fill_between(xx, sp.stats.beta(5, 3).pdf(xx))
plt.ylim(0, 6)
plt.title("          a=5 b=3 MODE={0}".format((5-1)/(5+3-2)))
plt.subplot(223)
plt.fill_between(xx, sp.stats.beta(7, 4).pdf(xx))
plt.ylim(0, 6)
plt.title("a=7 b=4 MODE={0}          ".format((7-1)/(7+4-2)))
plt.subplot(224)
plt.fill_between(xx, sp.stats.beta(9, 5).pdf(xx))
plt.ylim(0, 6)
plt.title("          a=9 b=5 MODE={0}".format((9-1)/(9+5-2)))
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def plot_triangle(X, kind):
    n1 = np.array([10, 0, 0])
    n2 = np.array([0, 5, 0])
    n3 = np.array([0, 0, 3])
    n12 = (n1 + n2) / 2
    m1 = np.array([1, -1, 0])
    m2 = n3 - n12
    m1 = m1 / np.linalg.norm(m1)
    m2 = m2 / np.linalg.norm(m2)

    X1 = (X - n12).dot(m1)
    X2 = (X - n12).dot(m2)

    sns.jointplot(X1, X2, kind=kind,  xlim=(-2.5,-1), ylim=(4, 5.5))
    plt.show()
```

```python
np.random.seed(0)
X1 = np.random.rand(1000, 3)
X1 = X1 / X1.sum(axis=1)[:, np.newaxis]
plot_triangle(X1, kind="scatter")
```

```python
plot_triangle(X1, kind="hex")
```

```python
X2 = sp.stats.dirichlet((0.2, 0.3, 0.5)).rvs(1000)
plot_triangle(X2, kind="scatter")
```

```python
plot_triangle(X2, kind="hex")
```
