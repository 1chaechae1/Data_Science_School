```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0)
n = 5
mu = 0.5
rv = sp.stats.binom(n, mu)
x = rv.rvs(1000)
sns.countplot(x)
plt.title("Simulation Result of Binomial Variable")
plt.xlabel("Sample Value")
plt.show()
```

```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def plot_binom(mu, N, M):
    rv = sp.stats.binom(N, mu)
    x = rv.rvs(M, random_state=0)
    y = np.bincount(x, minlength=N+1) / float(len(x))
    xx = np.arange(N+1)
    df = pd.DataFrame({'Theory': rv.pmf(xx), 'Simulation': y})
    df = pd.DataFrame({'Theory': rv.pmf(xx), 'Simulation': y}).stack()
    df = df.reset_index()
    df.columns = ['sample value', 'type', 'ratio']
    df.pivot('sample value', 'type', 'ratio') 
    sns.barplot(x='sample value', y='ratio', hue='type', data=df)
    plt.title('$\mu$={}, $N$={}, sample_size={}'.format(mu, N, M))
    plt.ylim(0, 0.7)

plt.subplot(221)
plot_binom(0.5, 5, 10)
plt.subplot(222)
plot_binom(0.5, 5, 1000)
plt.subplot(223)
plot_binom(0.9, 20, 10)
plt.subplot(224)
plot_binom(0.9, 20, 1000)
plt.tight_layout()
plt.show()
```
