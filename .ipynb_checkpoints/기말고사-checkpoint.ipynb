```python
N=20
mu=0.5
rv=sp.stats.binom(N,mu)
1-rv.cdf(11-1)
```

```python
xx = np.arange(N+1)

plt.subplot(211)
plt.stem(xx, rv.pmf(xx))
plt.ylabel("pmf")
plt.title("pmf of the test statistic distribution")
black = {"facecolor": "black"}
plt.annotate('Test Statistics t=11', xy=(11, 0.0), xytext=(11, 0.4), arrowprops=black)

plt.subplot(212)
plt.stem(xx, rv.cdf(xx))
plt.ylabel("cdf")
plt.title("cdf of the test statistic distribution")
plt.annotate('Test Statistics t=11', xy=(11,0.0), xytext=(11, 0.3),arrowprops=black)

plt.tight_layout()
plt.show()
```

```python
N1=10
pvalues1=np.array([sp.stats.binom_test(n, N1) for n in range(N1+1)])
fair_range1, =np.where(pvalues1>0.1)
fair_start1, fair_end1 = fair_range1[0], fair_range1[-1]
print('If N={} is fair: When the number of heads is {}~{}'.format(N1, fair_start1, fair_end1))

N2=1000
pvalues2=np.array([sp.stats.binom_test(n, N2) for n in range(N2+1)])
fair_range2, =np.where(pvalues2>0.1)
fair_start2, fair_end2 = fair_range2[0], fair_range2[-1]
print('If N={} is fair: When the number of heads is {}~{}'.format(N2, fair_start2, fair_end2))

plt.subplot(211)
plt.plot(pvalues1, 'r-')
plt.hlines(y=0.1, xmin=0, xmax=N1+1, ls='--')
plt.xlim(0,N1)
plt.title('Significant probability when N={}'.format(N1))
plt.subplot(212)
plt.plot(pvalues2, 'r-')
plt.hlines(y=0.1, xmin=0, xmax=N2+1, ls='--')
plt.xlim(0,N2)
plt.title('Significant probability when N={}'.format(N2))
plt.xlabel("The number comes up heads")
plt.tight_layout()
plt.show()
```
