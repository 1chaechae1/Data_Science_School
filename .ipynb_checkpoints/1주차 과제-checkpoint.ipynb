```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
import matplotlib.font_manager
```

```python
y = [300, 400, 700, 300, 100, 100, 100]
x = np.arange(len(y))
xlabel = [0, 1, 2, 3, 4, 5, 6]
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("Number of childern")
plt.ylabel("Frequency")
plt.show()
```

```python
x=np.array([0, 1, 2, 3, 4, 5, 6])
y=[0.15, 0.20, 0.35, 0.15, 0.05, 0.05, 0.05]
hist=plt.hist(x, bins = 6, weights = y, density=False, cumulative=False)
plt.title("Histogram")
plt.show()
```

```python
names = [0, 1, 2, 3, 4, 5, 6]
scores = [300, 400, 700, 300, 100, 100, 100]
colors = ['yellowgreen', 'lightcoral', 'skyblue', 'gold', 'lightskyblue', 'lightyellow', 'lightgreen']
explode = (0, 0, 0.1, 0, 0, 0, 0)

plt.pie(scores, explode=explode,
       labels=names, colors=colors,
       autopct='%1.2f%%', shadow=True,
       startangle=90)

plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```
