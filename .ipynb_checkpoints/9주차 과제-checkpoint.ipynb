```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import scipy.stats
import seaborn as sns

np.random.seed(0)

rv = sp.stats.norm()
M = 5000

plt.subplot(1, 2, 1)
N = 2
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 800)
plt.title("distri. of sum of square (N = 2)")
plt.xlabel("sample value")
plt.ylabel("number")
plt.show()
```

```python
plt.subplot(1, 2, 2)
N = 30
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 800)
plt.title("distri. of sum of square (N = 30)")
plt.xlabel("sample value")
plt.show()
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 10000

N1 = 2
x1 = rv.rvs((N1, M))
t1 = (x1 ** 2).sum(axis=0)

N2 = 2
x2 = rv.rvs((N2, M))
t2 = (x2 ** 2).sum(axis=0)

t = t2 / t1
sns.distplot(t, bins=200, kde=False)
plt.axvline(1, ls=":");
plt.xlim(-0.1, 3)
plt.title("ratio of sum of squares (N1=2, N2=2)")
plt.xlabel("sample value")
plt.ylabel("sample number")
plt.show()
```
