```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt

xx = np.linspace(0, 1, 1000)
aA, bA = 3, 2
plt.plot(xx, sp.stats.beta(aA, bA).pdf(xx), c='r', ls='-', label='Parameters of Product A')
aB, bB = 61, 41
plt.plot(xx, sp.stats.beta(aB, bB).pdf(xx), c='g', ls='--', label='Parameters of Product B')
plt.legend()
plt.title('dist. of Bernouli dist. parameter calculated by Bayesian estimation method')
plt.show()
```
